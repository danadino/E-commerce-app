```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Step 2: Load the datasets
customers_df = pd.read_csv('E-commerce_data - customers.csv')
genders_df = pd.read_csv('E-commerce_data - genders.csv')
cities_df = pd.read_csv('E-commerce_data - cities.csv')
transactions_df = pd.read_csv('E-commerce_data - transactions.csv')
branches_df = pd.read_csv('E-commerce_data - branches.csv')
merchants_df = pd.read_csv('E-commerce_data - merchants.csv')

# Step 3: Inspect the first few rows of each DataFrame to understand their structure
print(customers_df.head())
print(genders_df.head())
print(cities_df.head())
print(transactions_df.head())
print(branches_df.head())
print(merchants_df.head())

# Step 4: Merge the customer data with gender and city information
customers_with_demographics = customers_df.merge(genders_df, on='gender_id', how='left') \
                                          .merge(cities_df, on='city_id', how='left')
print(customers_with_demographics.head())

# Step 5: Merge transactions with customer demographics
transactions_with_customer_info = transactions_df.merge(customers_with_demographics, on='customer_id', how='left')
print(transactions_with_customer_info.head())

# Step 6: Merge transactions with branch and merchant data
transactions_with_branch_info = transactions_with_customer_info.merge(branches_df, on='branch_id', how='left') \
                                                              .merge(merchants_df, on='merchant_id', how='left')
print(transactions_with_branch_info.head())

# Step 7: Analyze Coupon Usage over Time (Claimed vs. Burnt)
transactions_with_branch_info['transaction_date'] = pd.to_datetime(transactions_with_branch_info['transaction_date'])
coupon_usage_over_time = transactions_with_branch_info.groupby([transactions_with_branch_info['transaction_date'].dt.to_period('M'), 'transaction_status']).size().unstack().fillna(0)

# Plot Coupon Usage over Time
plt.figure(figsize=(10, 6))
coupon_usage_over_time.plot(kind='line', title='Coupon Usage Over Time')
plt.ylabel('Count')
plt.xlabel('Month')
plt.show()

# Step 8: Analyze Branch Performance (Based on Coupon Burns)
branch_performance = transactions_with_branch_info[transactions_with_branch_info['transaction_status'] == 'burnt'].groupby('branch_id').size()
branch_performance = branch_performance.sort_values(ascending=False)

# Plot Top-performing Branches
plt.figure(figsize=(10, 6))
sns.barplot(x=branch_performance.index, y=branch_performance.values)
plt.title('Top-performing Branches')
plt.xlabel('Branch ID')
plt.ylabel('Number of Burnt Coupons')

# Step 9: Customer Retention - Calculate the Number of Coupons Claimed by Each Customer
customer_retention = transactions_with_branch_info.groupby('customer_id').size()

# Plot Customer Retention
plt.figure(figsize=(10, 6))
sns.histplot(customer_retention, kde=True)
plt.title('Customer Retention and Loyalty')
plt.xlabel('Number of Coupons Claimed')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 1: Feature Selection for Segmentation
# Extract relevant features (Coupon Claimed Count and Coupon Burned Count)
customer_coupon_counts = transactions_with_branch_info.groupby('customer_id').agg({
    'transaction_id': 'count',  # Count of transactions (claimed)
    'burn_date': 'count'  # Count of burnt coupons
}).rename(columns={'transaction_id': 'coupon_claimed_count', 'burn_date': 'coupon_burned_count'}).reset_index()

# Step 2: Standardize the Features (Important for clustering algorithms like K-Means)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_coupon_counts[['coupon_claimed_count', 'coupon_burned_count']])

# Step 3: Explore Different Numbers of Clusters (K) from 2 to 10
inertia = []
silhouette_scores = []

# Trying different values of K (clusters) from 2 to 10 to evaluate model performance
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(scaled_features, kmeans.labels_))

# Step 4: Plot Inertia vs Number of Clusters (Elbow Method)
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), inertia, marker='o', linestyle='-', color='b', label='Inertia')
plt.title('Inertia vs Number of Clusters (Elbow Method)')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# Step 5: Plot Silhouette Scores vs Number of Clusters
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='-', color='g', label='Silhouette Score')
plt.title('Silhouette Score vs Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Step 6: Choose Optimal Number of Clusters Based on Elbow and Silhouette Score
# Based on the plots, choose the optimal number of clusters
optimal_k = 4  # This will be determined based on your plots (choose the optimal k)

# Step 7: Fit the K-Means Model with the Chosen Number of Clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
customer_coupon_counts['customer_segment'] = kmeans.fit_predict(scaled_features)

# Step 8: Evaluate Clustering Using Silhouette Score and Inertia
silhouette = silhouette_score(scaled_features, customer_coupon_counts['customer_segment'])
inertia_value = kmeans.inertia_

print(f"Optimal Number of Clusters: {optimal_k}")
print(f"Silhouette Score for {optimal_k} clusters: {silhouette}")
print(f"Inertia for {optimal_k} clusters: {inertia_value}")

# Step 9: Visualize the Customer Segments
plt.figure(figsize=(10, 6))
plt.scatter(customer_coupon_counts['coupon_claimed_count'], customer_coupon_counts['coupon_burned_count'],
            c=customer_coupon_counts['customer_segment'], cmap='viridis', s=50)
plt.xlabel('Coupon Claimed Count')
plt.ylabel('Coupon Burned Count')
plt.title(f'Customer Segmentation (K={optimal_k})')
plt.colorbar(label='Segment')
plt.show()

# Step 10: Analyze and Describe Each Segment
segment_analysis = customer_coupon_counts.groupby('customer_segment').agg({
    'coupon_claimed_count': ['mean', 'std'],
    'coupon_burned_count': ['mean', 'std']
})

# Display Segment Analysis
print("\nSegment Analysis:")
print(segment_analysis)

# Step 11: Make Recommendations for Each Segment
for segment, data in segment_analysis.iterrows():
    print(f"\nSegment {segment}:")
    print(f" - Average Coupons Claimed: {data['coupon_claimed_count']['mean']}")
    print(f" - Average Coupons Burned: {data['coupon_burned_count']['mean']}")
    print(f"Recommendation: Tailor marketing campaigns and coupon offerings to this group based on their behavior.")
```
